```python
usuarios=[]
seleccion=[]
datos=data['5063'].split(',')
for i in range(int(np.floor(len(data['5063'].split(',')) /5))):
    usuarios.append(10000 +i)
    seleccion.append(datos[i*5:i*5+5])
estudio=pd.DataFrame(zip(usuarios,seleccion),columns=['usuario','seleccion'])
for i in range(1,6):
    estudio[f'sel_{i}']=estudio['seleccion'].apply(lambda x: x[i-1])
estudio.drop('seleccion',axis=1,inplace=True)
estudio.to_csv('Informacion_estudio.csv',index=False)
```

```python
data_mege=pd.merge(apartamentos,comentarios,left_on='URL',right_on='apart_id')
```

```python
# Crear la matriz usuario-item usando pivot_table
matriz_usuario_item = data_mege.pivot_table(index='user_id', columns='ID', values='rating', fill_value=0)
matriz_usuario_item
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
item_encoder = LabelEncoder()
comentarios['apart_id'] = item_encoder.fit_transform(comentarios['apart_id'])

user_encoder = LabelEncoder()
comentarios['user_id'] = user_encoder.fit_transform(comentarios['user_id'])

X_train, X_test, y_train, y_test = train_test_split(comentarios[['user_id', 'apart_id']], comentarios['rating'], test_size=0.2, random_state=42)
```

```python
num_users = comentarios['user_id'].nunique()
num_items = apartamentos['ID'].nunique()
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Dense, Concatenate, Flatten
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.regularizers import l2# Definir tamaño de los embeddings
embedding_size = 2
# Entradas del modelo
user_id_input = Input(shape=(1,), name='user_id_input')
item_id_input = Input(shape=(1,), name='item_id_input')

# Embeddings de usuario y apartamento
user_embedding = Embedding(input_dim=num_users, output_dim=embedding_size, input_length=1, name='user_embedding')(user_id_input)
item_embedding = Embedding(input_dim=num_items, output_dim=embedding_size, input_length=1, name='item_embedding')(item_id_input)

# Aplanar embeddings
user_vecs = Flatten()(user_embedding)
item_vecs = Flatten()(item_embedding)

# Concatenar embeddings
concat = Concatenate()([user_vecs, item_vecs])

# Capas densas para la predicción final
dense1 = Dense(10, activation='relu')(concat)
dense2 = Dense(10, activation='relu')(dense1)
output = Dense(1, activation='sigmoid', name='output')(dense2)

# Modelo final
model = Model(inputs=[user_id_input, item_id_input], outputs=output)

# Compilar el modelo
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])

# Resumen del modelo
model.summary()
```

```python
# Entrenamiento del modelo
history = model.fit([X_train['user_id'], X_train['apart_id']], y_train,
                    batch_size=2,
                    epochs=10,
                    validation_data=([X_test['user_id'], X_test['apart_id']], y_test))
```
