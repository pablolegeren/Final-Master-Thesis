```python
from deepface import DeepFace
```

```python
objs = DeepFace.analyze('https://a0.muscache.com/im/pictures/user/254d768d-bd9c-49d5-82d8-9842bee9ddd4.jpg?im_w=240')
objs
```

```python
objs[0]['age']
```

```python
def analyze_image(image_path):
    try:
        result = DeepFace.analyze(image_path)
        edad = result[0]['age']
        genero = result[0]['dominant_gender']
    except:
        edad = None
        genero = None
    return edad, genero

fotos[['edad', 'genero']] = fotos['imagen'].apply(lambda x: pd.Series(analyze_image(x)))
```
